```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
CHESTNUT_BUN_DIAMETER = 0.1  # 10 cm in meters
SOLAR_SYSTEM_RADIUS = 4.5e12  # 4.5 billion kilometers in meters
DOUBLING_TIME = 5 * 60  # 5 minutes to seconds

# Calculate initial volume of the chestnut bun
def calculate_volume(diameter):
    radius = diameter / 2
    return (4/3) * np.pi * (radius ** 3)

# Calculate the volume of the solar system
solar_system_volume = calculate_volume(SOLAR_SYSTEM_RADIUS)  # Correct radius

# Function to calculate the time to fill a given volume
def time_to_fill(initial_volume, target_volume, doubling_time):
    time = 0
    volumes = [initial_volume]
    times = [time]
    
    while volumes[-1] < target_volume:
        time += doubling_time
        volumes.append(volumes[-1] * 2)
        times.append(time)
    
    return times, volumes

# Initial volume of the chestnut bun
initial_volume = calculate_volume(CHESTNUT_BUN_DIAMETER)

# Calculate the time and volumes
times, volumes = time_to_fill(initial_volume, solar_system_volume, DOUBLING_TIME)

# Convert times to hours for better readability
times_hours = np.array(times) / 3600

# Identify the exact time when the solar system is filled
solar_system_fill_time = times_hours[-1]
solar_system_fill_volume = solar_system_volume


plt.figure(figsize=(10, 6))
plt.plot(times_hours, volumes, marker='o', label='Chestnut Bun Volume')
plt.axhline(y=solar_system_volume, color='r', linestyle='--', label='Volume of Solar System')

# Highlight the point where the solar system is filled
plt.scatter([solar_system_fill_time], [solar_system_fill_volume], color='red', label='Solar System Filled', zorder=5)


plt.xlabel('Time (hours)')
plt.ylabel('Volume (mÂ³)')
plt.title('Exponential Growth of Chestnut Bun Volume Over Time')
plt.legend()
plt.grid(True)
plt.show()


print(f"Time to fill the solar system: {solar_system_fill_time:.2f} hours")
```
