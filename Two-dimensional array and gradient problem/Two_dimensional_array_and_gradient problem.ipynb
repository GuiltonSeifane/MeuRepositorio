```python
import numpy as np

# Create x array
x = np.arange(-50, 50.1, 0.1)

# Compute y
y = 12 * x + 1

print("x values:", x)
print("y values:", y)
```

```python
array_xy = np.column_stack((x, y))
print("Combined array (x, y):", array_xy)
```

```python
# Calculate the gradient
gradient = np.diff(y) / np.diff(x)
gradient
```

```python
import matplotlib.pyplot as plt

# Plot the linear function
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, y, label='y = 12x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
plt.legend()

# Plot the gradient
plt.subplot(1, 2, 2)
plt.plot(x[:-1], gradient, label='Gradient', color='orange')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.title('Gradient of Linear Function')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.

    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    x = np.arange(*x_range)
    y = function(x)
    array_xy = np.column_stack((x, y))
    gradient = np.diff(y) / np.diff(x)
    return array_xy, gradient

def function1(array_x):
    return 12 * array_x + 1

def function2(array_x):
    return array_x**2

def function3(array_x):
    return 2 * array_x**2 + 2 * array_x

def function4(array_x):
    return np.sin(array_x**2)

# Compute gradients for different functions
array_xy1, gradient1 = compute_gradient(function1)
array_xy2, gradient2 = compute_gradient(function2)
array_xy3, gradient3 = compute_gradient(function3)
array_xy4, gradient4 = compute_gradient(function4)

# Plot the functions and their gradients
plt.figure(figsize=(18, 12))

# Function 1
plt.subplot(2, 2, 1)
plt.plot(array_xy1[:, 0], array_xy1[:, 1], label='y = 12x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(array_xy1[:-1, 0], gradient1, label='Gradient', color='orange')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.title('Gradient of Linear Function')
plt.legend()

# Function 2
plt.subplot(2, 2, 3)
plt.plot(array_xy2[:, 0], array_xy2[:, 1], label='y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(array_xy2[:-1, 0], gradient2, label='Gradient', color='orange')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.title('Gradient of Quadratic Function')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Function 2: y = x^2
min_y2 = np.min(array_xy2[:, 1])
min_index2 = np.argmin(array_xy2[:, 1])
min_x2 = array_xy2[min_index2, 0]

print(f"Minimum value of y for y = x^2: {min_y2} at x = {min_x2}")

# Function 3: y = 2x^2 + 2x
min_y3 = np.min(array_xy3[:, 1])
min_index3 = np.argmin(array_xy3[:, 1])
min_x3 = array_xy3[min_index3, 0]

print(f"Minimum value of y for y = 2x^2 + 2x: {min_y3} at x = {min_x3}")

# Function 4: y = sin(x^2)
min_y4 = np.min(array_xy4[:, 1])
min_index4 = np.argmin(array_xy4[:, 1])
min_x4 = array_xy4[min_index4, 0]

print(f"Minimum value of y for y = sin(x^2): {min_y4} at x = {min_x4}")
```
