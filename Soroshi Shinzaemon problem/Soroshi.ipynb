```python
import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days=100):
    """
    A function to calculate the rice grains received daily and the cumulative total
    based on the Sorori Shinzaemon anecdote.

    Parameters
    ----------
    n_days: int
        Number of days for which the calculation is done (default: 100).

    Returns
    -------
    list_n_grains : list
        List of rice grains received each day (length equals n_days).
    list_total_grains : list
        List of cumulative rice grains received up to each day (length equals n_days).
    """
    list_n_grains = [2**i for i in range(n_days)]
    list_total_grains = [sum(list_n_grains[:i+1]) for i in range(n_days)]
    return list_n_grains, list_total_grains
```

```python
list_n_grains, list_total_grains = compute_sorori_shinzaemon(100)

print("Number of rice grains on the 100th day:", list_n_grains[-1])
print("Total rice grains received by the 100th day:", list_total_grains[-1])

days = list(range(1, 101))
plt.figure(figsize=(12, 6))
plt.plot(days, list_n_grains, label="Rice grains per day", color="blue")
plt.plot(days, list_total_grains, label="Total rice grains", color="orange")
plt.title("Rice Grains Received Over Time")
plt.xlabel("Days")
plt.ylabel("Number of Rice Grains")
plt.legend()
plt.yscale('log')  # Use log scale to handle large numbers
plt.grid(True)
plt.show()
```

```python
def numbRice (days):
    
    list_n_grains, list_total_grains = compute_sorori_shinzaemon(days)

    print(f"Number of rice grains on day {days}:", list_n_grains[-1])
    print(f"Total rice grains received by day {days}:", list_total_grains[-1])

    plt.figure(figsize=(12, 6))
    plt.plot(range(1, days + 1), list_n_grains, label="Rice grains per day", color="green")
    plt.plot(range(1, days + 1), list_total_grains, label="Total rice grains", color="red") 
    plt.title(f"Rice Grains Over {days} Days")
    plt.xlabel("Days")
    plt.ylabel("Number of Rice Grains")
    plt.legend()
    plt.yscale('log')  # Use log scale to handle large numbers
    plt.grid(True)
    plt.show()
numbRice(81)
```

```python
def calculate_days_of_life(total_grains, num_people):
    """
    To alculate how many days a given number of people can survive with the provided rice grains.

    Parameters
 
    total_grains : int
        Total number of rice grains available.
    num_people : int
        Number of people sharing the rice.

    Returns
  
    int
        Number of days the given people can survive.
    """
    grains_per_day_per_person = 25000  # Grains of rice needed per person per a day
    total_daily_consumption = num_people * grains_per_day_per_person
    return total_grains // total_daily_consumption

# Example: 81 people and total grain received in 100 days
num_people = 81
total_grains = list_total_grains[-1]
days_supported = calculate_days_of_life(total_grains, num_people)

print(f"With {total_grains} rice grains, {num_people} people can live for {days_supported} days.")
```
