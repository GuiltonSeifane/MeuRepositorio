```python
import matplotlib.pyplot as plt

initial_thickness = 0.00008  
folds = 43  

thicknesses = [initial_thickness]  

# Calculate thicknesses for each fold
for i in range(1, folds + 1):
    new_thickness = thicknesses[-1] * 2   
    thicknesses.append(new_thickness)  

# Customize the graph
plt.title("Thickness of Folded Paper", fontsize=16)  
plt.xlabel("Number of Folds", fontsize=14) 
plt.ylabel("Thickness [m]", fontsize=14)  

# Customizations:

plt.plot(thicknesses, color='green', linestyle=':', linewidth=3)


plt.tick_params(labelsize=12)


plt.grid(True)
plt.show()
```
