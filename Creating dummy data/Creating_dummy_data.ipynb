```python
import numpy as np

# Fixing the seed for reproducibility
np.random.seed(0)

# Defining the mean and covariance matrix
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]

#Generating 500 random numbers
data_1 = np.random.multivariate_normal(mean, cov, 500)

# Check the shape
print(data_1.shape)  # Output: (500, 2)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot
plt.scatter(data_1[:, 0], data_1[:, 1], label="Problem 1 Data", alpha=0.6)
plt.title("Scatter Plot of Problem 1 Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create histograms for each dimension
plt.figure(figsize=(12, 5))

# Histogram for the first dimension (X-axis)
plt.subplot(1, 2, 1)
plt.hist(data_1[:, 0], bins=20, color='blue', alpha=0.7)
plt.title("Histogram of X-axis")
plt.xlabel("X-axis")
plt.ylabel("Frequency")
plt.xlim(-6, 0)  # Align the horizontal axis range

# Histogram for the second dimension (Y-axis)
plt.subplot(1, 2, 2)
plt.hist(data_1[:, 1], bins=20, color='green', alpha=0.7)
plt.title("Histogram of Y-axis")
plt.xlabel("Y-axis")
plt.ylabel("Frequency")
plt.xlim(-4, 4)  

plt.tight_layout()
plt.show()
```

```python
# Define the new mean
mean_2 = [0, -3]

# Generate 500 new random numbers
data_2 = np.random.multivariate_normal(mean_2, cov, 500)

# Visualize both datasets on the same scatter plot
plt.scatter(data_1[:, 0], data_1[:, 1], label="Problem 1 Data", alpha=0.6)
plt.scatter(data_2[:, 0], data_2[:, 1], label="Problem 4 Data", alpha=0.6)
plt.title("Scatter Plot of Problem 1 and Problem 4 Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Combining  the two datasets
combined_data = np.vstack((data_1, data_2))

# printing the shape
print(combined_data.shape)  # Output: (1000, 2)
```

```python
# Creating labels
labels = np.array([0] * 500 + [1] * 500)

# Add labels as a new column
labeled_data = np.hstack((combined_data, labels.reshape(-1, 1)))

# Check the shape
print(labeled_data.shape)  # Output: (1000, 3)

# Example of the resulting ndarray
print(labeled_data[:5])  # First 5 rows of the labeled data
```
